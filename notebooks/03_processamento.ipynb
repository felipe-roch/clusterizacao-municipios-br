```python
import pandas as pd
from pathlib import Path

caminho = Path(r"C:\Users\Felipe\Documents\Python\Clusterizacao_Municipios\dados")

atlas = pd.read_csv(caminho / 'processados' / 'Atlas_tratado.csv')
sidra = pd.read_csv(caminho / 'processados' / 'sidra_consolidado.csv')

display(atlas.head(10))
display(sidra.head(10))
display(atlas.info())
display(sidra.info())
```

```python
sidra_nomes = set(sidra['municipio'])
atlas_nomes = set(atlas['municipio_sidra'])

print(f'Municípios na Base do Atlas: {len(atlas_nomes)}')
print(f'Municípios na Base do Sidra: {len(sidra_nomes)}\n')

apenas_sidra = sidra_nomes - atlas_nomes
print(f"Municípios no SIDRA sem correspondência no Atlas: {len(apenas_sidra)}")
print(sorted(apenas_sidra))
```

```python
apenas_atlas = atlas_nomes - sidra_nomes
print(f"Municípios no Atlas sem correspondência no SIDRA: {len(apenas_atlas)}")
print(sorted(apenas_atlas))
```

```python
atlas['municipio_sidra_lower'] = atlas['municipio_sidra'].str.lower()
sidra['municipio_lower'] = sidra['municipio'].str.lower()

apenas_sidra_lower = set(sidra['municipio_lower']) - set(atlas['municipio_sidra_lower'])
print(f"Sem correspondência após lowercase: {len(apenas_sidra_lower)}\n")
print(sorted(apenas_sidra_lower))
```

```python
apenas_atlas_lower = set(atlas['municipio_sidra_lower']) - set(sidra['municipio_lower'])
print(f"Sem correspondência após lowercase: {len(apenas_atlas_lower)}\n")
print(sorted(apenas_atlas_lower))
```
