```python
import pandas as pd
from pathlib import Path

caminho = Path(r"C:\Users\Felipe\Documents\Python\Clusterizacao_Municipios\dados")

atlas = pd.read_excel(caminho / "brutos" / "Atlas_BR_Municipios.xlsx")

display(atlas.head())
display(atlas.shape)
display(atlas.info())
```

```python
display(atlas['Territorialidade'].head(10))
```

```python
atlas = atlas.drop(columns=[col for col in atlas.columns if "Posição" in col])
display(atlas.columns)
```

```python
atlas = atlas.rename(columns={'Territorialidade': 'municipio'})
display(atlas.head(10))
```

```python
atlas['municipio_sidra'] = atlas['municipio'].str.replace(' (', ' - ', regex=False).str.replace(')', '', regex=False)

display(atlas.head(10))
```

```python
atlas.to_csv(caminho / 'processados' / 'Atlas_tratado.csv', index=False)
```
