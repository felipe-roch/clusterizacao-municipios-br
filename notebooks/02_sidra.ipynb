```python
import sidrapy
import pandas as pd
from pathlib import Path

caminho = Path(r"C:\Users\Felipe\Documents\Python\Clusterizacao_Municipios\dados")

dados = sidrapy.get_table(
    table_code="6803",
    territorial_level="6",
    ibge_territorial_code="all",
    period="2022"
)

display(dados.head())
display(dados.info())
```

```python
dados.columns = dados.iloc[0]
dados = dados.iloc[1:]

display(dados.head())
```

```python
import requests

response = requests.get("https://servicodados.ibge.gov.br/api/v3/agregados/9923/metadados")
metadados = response.json()
display(metadados['classificacoes'])
```

```python
dados_domicilios = sidrapy.get_table(
    table_code="9923",
    territorial_level="6",
    ibge_territorial_code="all",
    period="2022",
    classifications={"1": "all"}
)

display(dados_domicilios.head())
display(dados_domicilios.info())
```

```python
dados_domicilios.columns = dados_domicilios.iloc[0]
dados_domicilios = dados_domicilios[1:]

dados_domicilios = dados_domicilios[dados_domicilios['Situação do domicílio (Código)'] != '2']

display(dados_domicilios.head())
```

```python
dom_total = dados_domicilios[dados_domicilios['Situação do domicílio (Código)'] == '6795'][['Município (Código)','Valor']]
dom_urb = dados_domicilios[dados_domicilios['Situação do domicílio (Código)'] == '1'][['Município (Código)','Valor']]

dom = pd.merge(dom_urb, dom_total, on='Município (Código)', how='left')

display(dom.head())
```

```python
dom['Valor_x'] = pd.to_numeric(dom['Valor_x'], errors='coerce')
dom['Valor_y'] = pd.to_numeric(dom['Valor_y'], errors='coerce')

dom['per_domicilio_urbano'] = (dom['Valor_x'].div(dom['Valor_y'])) * 100

dom = dom.drop(columns=['Valor_x', 'Valor_y'])

display(dom.head())
```

```python
dom = dom.rename(columns={'Município (Código)': 'cod_municipio'})

dados_consolidados = pd.merge(dados_consolidados, dom, on='cod_municipio', how='left')

display(dados_consolidados.head())
```

```python
dados_consolidados.to_csv(caminho / 'processados' / 'sidra_consolidado.csv', index=False)
```
